```python
import numpy as np
import pandas as pd
```

```python
# lendo url csv
url = 'https://raw.githubusercontent.com/trustinyoon/2020-NBA-Chip-Predictor/master/past_league_rankings.csv'
df_2003_to_2019 = pd.read_csv(url)
df_2003_to_2019.head()
```

```python
df_2020 = pd.read_csv('2020_final.csv')

# adiciona uma coluna com o ano
df_2020['Year'] = 2020

df_2020
```

```python
df_2021 = pd.read_csv('2021_final.csv')

# adiciona uma coluna com o ano
df_2021['Year'] = 2021

df_2021
```

```python
df_2022 = pd.read_csv('2022_final.csv')

# adiciona uma coluna com o ano
df_2022['Year'] = 2022

df_2022.head()
```

```python
df_2023 = pd.read_csv('2023_final.csv')

# adiciona uma coluna com o ano
df_2023['Year'] = 2023

df_2023.head()
```

```python
# junte os dataframes
df = pd.concat([df_2003_to_2019, df_2020, df_2021, df_2022, df_2023])
```

```python
salvar = df.to_csv('final.csv', index=False)


```
