```python
# transformando a coluna 'Playoff Wins' em int
df_advanced['Playoff Wins'] = df_advanced['Playoff Wins'].astype(int)

# ordenando os dados pela coluna 'Playoff Wins'
df_advanced = df_advanced.sort_values(by=['Playoff Wins'], ascending=False)

# apagando as coluna que contem apenas nan
df_advanced = df_advanced.dropna(axis=1, how='all')

# resetando o index e trasnformando o index em uma coluna chamada 'Rk'
df_advanced = df_advanced.reset_index()
df_advanced = df_advanced.drop(columns=['index'])

# criando a coluna 'Rk' e atribuindo o valor do index + 1
df_advanced['Rk'] = df_advanced.index + 1

df_advanced.head()
```

```python
colunas = df_advanced.columns

# remove as colunas que nÃ£o serÃ£o utilizadas 'Playoff Wins', 'Rk'
colunas = colunas.drop(['Playoff Wins', 'Rk', 'Rk_x', 'Rk_y'])

# remove as colunas repetidas 'Rk_x', 'Rk_y'
df_advanced = df_advanced.drop(columns=['Rk_x', 'Rk_y'])
```

```python
# transforma todos os dados numericos em rank comparando com os outros times da liga na temporada 2020 (1 = melhor time da liga, 16 = pior time da liga) para as colunas em 'colunas'

for coluna in colunas:
    if coluna != 'Team':
        df_advanced[coluna] = df_advanced[coluna].rank(ascending=False)

df_advanced.head()
```
